```python
import pandas as pd
import numpy as np
```

```python
anime = pd.read_csv('anime.csv')
```

```python
rating = pd.read_csv('rating.csv')
```

```python
anime.drop('rating',axis=1,inplace=True)
```

```python
anime_rating = anime.merge(rating,on='anime_id')#joining table on anime_id column
anime_rating
```

```python
weebs = anime_rating.groupby('user_id').count()['rating'].reset_index()
weebs
```

```python
weebs = weebs[weebs['rating']>200]['user_id']
weebs
```

```python
weebs = anime_rating[anime_rating['user_id'].isin(weebs)]
weebs
```

```python
weebs.drop_duplicates()
```

```python
pt = weebs.pivot_table(index='name',columns='user_id',values='rating')
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(pt)
```

```python
similarity
```

```python
def recommend(anime_name):
    index = np.where(pt.index==anime_name)[0][0]
    similar_anime = sorted(list(enumerate(similarity[index])),key=lambda x:x[1],reverse = True)[1:6]
    for i in similar_anime:
        print(pt.index[i[0]])
```

```python
recommend('Death Note')
```
